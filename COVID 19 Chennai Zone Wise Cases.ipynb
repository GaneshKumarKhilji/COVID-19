```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# To increase figure size
plt.rcParams["figure.figsize"] = (15,8)
```

```python
#importing data with pandas
dataset=pd.read_csv('Zone Wise Breakup Week 3.csv')
dataset2=pd.read_csv('Zone Wise Breakup Week 4.csv')
```

```python
#Converting it into Numpy arrays and storing in variables
x=dataset.iloc[:,0].values
y=dataset.iloc[:,1].values
a=dataset2.iloc[:,0].values
b=dataset2.iloc[:,1].values
```

```python
print(list(zip(y,b)))
```

```python
k=[]
for i,j in zip(y,b):
    if j==0:
        g=0
    elif i==0:
        g=j
    else:
        g=(j-i)/i
    k.append(round(g,2))
```

```python
k
```

```python
#Plotting Graph with Matplotlib
plt.title('COVID 19 Week 3 Vs Week 4 Analysis\n\nChennai',fontsize=20,fontweight='bold',pad=20)
plt.scatter(x,y,color='red')
plt.scatter(a,b,color='red')
plt.plot(x,y,label='Week 3 Confirmed Cases')
plt.plot(a,b,label='Week 4 Confirmed Cases')
plt.xlabel('Areas')
plt.ylabel('Confirmed Cases')
plt.legend() #To show legend
plt.xticks(rotation=90) #To rotate X axis label
#To show values inside graph nearby plots
for index,data in enumerate(y):
    plt.text(x=index , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=12))
for index,data in enumerate(b):
    plt.text(x=index , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=12))
plt.show()
```

```python
plt.title('COVID 19 Growth Rate\n\nChennai',fontsize=20,fontweight='bold',pad=20)
plt.xlabel('Areas')
plt.ylabel('Growth Rate')
plt.text(10,10.4,'______________________________________')
plt.text(10,10,'| Growth Rate = Present - Past / Past |')
plt.text(10,10,'______________________________________')
plt.plot(x,k,label='Growth Rate')
plt.scatter(x,k,color='blue')
plt.xticks(rotation=90)
plt.legend() #To show legend
for index,data in enumerate(k):
    plt.text(x=index , y =data+1 , s=f"{data}%" , fontdict=dict(fontsize=12))
```
